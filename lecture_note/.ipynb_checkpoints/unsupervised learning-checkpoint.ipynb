## 비지도 학습(unsupervised learning)이란 무엇인가?
[University of Turku](<https://www.utu.fi/fi>)의 Data Analysis and Knowledge Discovery 강의 내용을 정리한 노트이며, 개인적인 의견과 자료조사를 더했습니다.

### 지도 학습 vs 비지도 학습

[지도 학습](https://euneestella.github.io/data%20analysis%20and%20knowledge%20discovery/2019/12/11/supervised-learning/)과 비지도 학습의 근본적인 차이는 정답 데이터(label)의 유무이다.

지도 학습 : $$\left\{ { \left( x_{ 1 },{ y }_{ 1 } \right) ,\quad \left( { x }_{ 2 },{ y }_{ 2 } \right) ,\cdots ,\left( { x }_{ m },{ y }_{ m } \right)  } \right\} $$

비지도 학습 : $$ \left\{ \left( { x }_{ 1 },{ x }_{ 2 },\cdots { x }_{ m } \right)  \right\}  $$

데이터가 라벨링되어 있다면 지도 학습을 사용하고, 그렇지 않다면 비지도 학습을 수행한다. 지도 학습이 정답(label)을 이미 알고 있을 때 컴퓨터가 문제를 해결하는 것인 반면 비지도학습은 <u>사람도 알지 못하는 정답을 컴퓨터가 찾기 바라는 것</u>이다.    

비지도 학습의 대표적인 기법에는 <b>군집 분석(cluster analysis)</b>, <b>연관 규칙(association rules)</b>, <b>편차 분석(deviation analysis)</b>이 있다.

### 군집 분석(cluster analysis)
군집 분석은 비슷한 데이터 오브젝트끼리 군(cluster)을 만드는 분석 기법이다. 같은 군 안에서는 서로 비슷하며(거리가 가까우며) 다른 군끼리는 덜 비슷하다(거리가 멀다).

#### 계층적 군집(hierarchical clustering)
계층적 군집 기법은 군들을 <u>하나씩 쌓아 올리는</u> 방법이다. 
- <u>응집형 계층적 군집</u>(agglomerative hierarchical clustering, <b>bottom-up method</b>) : 처음에 각각의 데이터 오브젝트를 서로 다른 군에 넣은 다음, 데이터를 탐색하며 서로 가까운 군끼리 합치는 방식
- <u>분리형 계층적 군집</u>(divisive hierarchical clustering, <b>top-down method</b>) : 데이터 오브젝트들이 하나의 군집에 속한다고 하고, 데이터를 탐색하며 작은 군으로 나누는 방식


<center>source : http://rpubs.com/Evan_Jung/hierarchical_clustering </center>

#### 군 사이의 이질성을 측정하는 방법

<img src ="https://github.com/euneestella/utu_data_analysis/blob/master/lecture_note/img/measuring-dissimilarity.PNG?raw=true">

- <b>중심 연결법(centroid)</b> : 두 군집의 중심 간 거리를 군집 간 거리로 정의
- <b>최단 연결법(single linkage)</b> : 두 군집의 최단 거리를 군집 간 거리로 정의 (가장 비슷한 데이터 오브젝트끼리)
- <b>최장 연결법(complete linkage)</b> : 두 군집의 최장 거리를 군집 간 거리로 정의 (가장 이질적인 데이터 오브젝트끼리)

#### k-평균 군집(k-Means clustering)

$$ \sum _{ i=1 }^{ k }{ \sum _{ x\in { C }_{ i } }^{  }{ \left\| x-{ p }_{ i } \right\| ^{ 2 } }  }  $$

${ C }_{ i }$가 i번째 군집이고, $p_{ i }$가 그 군집의 중심이라고 하자. 데이터 오브젝트 ${x}$과 $p_{ i }$ 사이의 거리의 제곱합을 함수로 정의(군 내부의 분산)하고 이를 최소화한다. 즉, <u>군 내부의 분산을 최소화하</u>도록 하는 군집화 알고리즘을 수행하는 것이다.

계층적 군집과 다른 점은 <u>군집의 수 k를 미리 정한다</u>는 점이다. <b>무작위하게 k개의 데이터 포인터를 골라 가장 가까운 군집의 중심</b>으로 삼는다. 이후 할당된 데이터 포인트들의 평균 벡터로 <b>새로운 군집의 중심(일종의 '무게중심')을 계산</b>한다. 

### 연관 규칙(association rule)

연관 규칙 학습은 경영학 방법론인 <u>장바구니 분석(market basket analysis)</u>을 위해 고안된 방식이다. 'A 고객이 B 상품과 함께 어떤 상품 C, D를 <b>함께</b> 구매하는지'를 분석한다.

단, B → C (If B then C) 는 C → B를 의미하지 않는다(<b>역은 성립하지 않는다</b>)는 점에 유의해야 한다.

#### 지지도(support)
$$P(B\cap C)$$
전체 거래 중 상품 B와 C를 함께 구매하는 비율이다. 
#### 신뢰도(confidence)
$$P\left( C|B \right) =\frac { P\left( B\cap C \right)  }{ P(B) } $$
상품 B를 포함하는 거래 중 상품 B와 C가 동시에 거래되는 비율이다.
#### 향상도(lift)
$$ \frac { P\left( C|B \right)  }{ P\left( C \right)  }  $$
B → C에서 임의로 C가 구매되는 경우 B가 고려되어 함께 구매되는 경우의 비율이다. 향상도가 <b>1인 경우</b> 두 상품이 <u>독립</u>적인 관계에 있음을 의미하며, <b>1보다 크면</b> <u>양의 관계</u>, <b>작으면</b> <u>음의 관계</u>에 있음을 의미한다.

### 편차 분석(deviation analysis)
연관 규칙 분석을 하다 보면, 비슷한 특성을 공유하고 있는 부분군을 발견할 수 있다. 예컨대 일반 고객보다 특정 상품을 많이 사는 고객들이 한 부분군이 될 수 있으며, 편차 분석은 부분군을 발견하는 분석 기법이다.
