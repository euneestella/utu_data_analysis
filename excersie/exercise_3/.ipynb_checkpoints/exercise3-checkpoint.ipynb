## Classification accuracy using leave-one-out cross validation

Again, predict the **ship type** using **speed, destination, transformed length, and breadth** of the ship as features. Find an estimation for the classification accuracy using *leave-one-out cross validation*. <br>
 - Use kNN classifier with k=5
     - What is the classification accuracy? Compare the result with the one you got in the previous task
     - Which method gives better evaluation of the performance of the classifier with this data set? Explain your choice

````python
loo = LeaveOneOut()
loo.get_n_splits(X_array)
````

Split array using ```LeaveOneOut```

````python
X_array = pd.DataFrame(X_array).to_numpy()
Y_array = pd.DataFrame(Y_array).to_numpy()
````

Turn from DataFrame to numpy array

````python
loocv_accuracy = []
````

Make empty list to store accuracy

````python
for train_index, test_index in loo.split(X_array) :
    X_train, X_test = X_array[train_index], X_array[test_index]
    Y_train, Y_test = Y_array[train_index], Y_array[test_index]
    knn.fit(X_train, Y_train.ravel())
    score = knn.score(X_test, Y_test)
    loocv_accuracy.append(score)
print(np.mean(loocv_accuracy))
````

Split train set and test set by using ```loo.split```.   

Then use ```knn.fit``` for kNN classifing the data. Store the accuracy with function ```knn.score``` and append the score into pre-made list ```loocv_accuracy_df```.   

Finally calculate the mean of our accuracy list.

## Model selection with leave-one-out cross validation

Select the best model (kNN with selection of k) by repeating the calculation of classification accuracy using leave-one-out cross validation with values k=1...20.
- Which value of k produces the best classification accuracy?
- Can you say something about the performance of this selected model with new, unseen data? Explain, how could you estimate the performance of this selected model?

````python
accuracy_df = pd.DataFrame(columns = ['k', 'accuracy'])
for i in range(20) :
    score_list = []
    for train_index, test_index in loo.split(X_array) :
        X_train, X_test = X_array[train_index], X_array[test_index]
        Y_train, Y_test = Y_array[train_index], Y_array[test_index]
        knn = knn = KNeighborsClassifier(n_neighbors = i+1)
        knn.fit(X_train, Y_train.ravel())
        score = knn.score(X_test, Y_test)
        score_list.append(score)
    accuracy_df = accuracy_df.append(pd.DataFrame([[i+1, np.mean(score_list)]], columns = ['k', 'accuracy']))
    accuracy_df = accuracy_df.reset_index(drop = True)
````

````python
accuracy_df
````

````python
print(accuracy_df['accuracy'].idxmax())
````

We can see the best performance when k = 5, which stands for index number 4
